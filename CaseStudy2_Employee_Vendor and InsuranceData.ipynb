# Task 6

```python
# find maximum date
view.TXN_DATE_TIME.max()
```

```python
#save maximumdate as current date
todays_date = pd.to_datetime('2021-06-30')
todays_date
```

```python
# show the values of columns 'TXN_DATE-TIME' and converting it into a datetime form
view['TXN_DATE_TIME'] = pd.to_datetime(view['TXN_DATE_TIME'])
view['TXN_DATE_TIME']
```

```python
# Fetch last 30 days value from column 'TXN_DATE_TIME'
last_30 = view.loc[(view['TXN_DATE_TIME'] > '2021-05-30') & (view['TXN_DATE_TIME'] < '2021-06-30')] 
last_30
```

```python
# avarage claim of each insurance type
avg_claim=last_30.groupby('INSURANCE_TYPE').mean()['CLAIM_AMOUNT'].to_dict()
avg_claim
```

```python
view
```

```python
# define a function for calculating claim_deviation
def Claim_deviation(row):
    if (todays_date - row['TXN_DATE_TIME'])<pd.Timedelta(days=30):
        return -1
    claim_deviation = avg_claim[row['INSURANCE_TYPE']] / row['CLAIM_AMOUNT']
    if claim_deviation < 0.5:
        return 1
    else:
        return 0
```

```python
# Applying the function to the dataframe view
view['CLAIM_DEVIATION'] = view.apply(Claim_deviation,axis=1)
view['CLAIM_DEVIATION']
```

```python
# value count of each 0 and 1 CLAIM_DEVIATION
view['CLAIM_DEVIATION'].value_counts()
```

```python
# Calculating total approved claims insurance type wise in last 3 days
claims_in_last_3_days = pd.DataFrame.from_dict(last_3_days[last_3_days['CLAIM_STATUS']=='A'].groupby('INSURANCE_TYPE').size().to_dict(),orient='index',columns=['NO_OF_CLAIMS_IN_LAST3DAYS'])
claims_in_last_3_days
```
