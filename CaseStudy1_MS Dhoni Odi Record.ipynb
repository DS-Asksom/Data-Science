# Data cleaning

# Task 1 : To  remove the first 2 characters from the opposition string because that is not required.

```python
data['opposition']=data['opposition'].apply(lambda x:x[2:])  # to remove v from opposition cols use lambda function
data['opposition']
```

```python
data   #all v removed
```

# Task 2 :Create a column indicating whether Dhoni was not out in that innings or not.

```python
import numpy as np
data['Not_out']=np.where(data['score'].str.endswith("*"),1,0)  #if true print 1 otherwise 0
```

```python
data
```

# Task 3: Drop the ODI number column because it is not required.

```python
data.drop(columns='odi_number',inplace=True) #inplace
data
```

# Task 4 : Drop all those matches from our records where Dhoni did not bat, and store this information in a new DataFrame.

```python
df=data.loc[((data['score']!='DNB') & (data['score'] !='TDNB'))]
df
```

```python
df.info()
```

# Task 5: Fix the data types of all the columns present in our new DataFrame.

```python
df['runs_scored']=df['runs_scored'].astype(int)#change datatype
df['balls_faced']=df['balls_faced'].astype(int) 
df['strike_rate']=df['strike_rate'].astype(float) 
df['fours']=df['fours'].astype(int) 
df['sixes']=df['sixes'].astype(int) 
```

```python
df.info()
```

```python
df['year']=pd.DatetimeIndex(df['date']).year
```
