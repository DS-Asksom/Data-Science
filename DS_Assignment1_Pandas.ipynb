```python
import pandas as pd
student_data1=pd.DataFrame({"Student_id":[0,1,2,3],"Name":['ajay','atul','akash','sanjay'],"Marks":[210,190,222,199]})
student_data1.set_index("Student_id",inplace=True)
print(student_data1)
```

```python
import pandas as pd
student_data2=pd.DataFrame({"Student_id":[0,1,2,3],"Name":['ashmita','ankita','rupali','rakesh'],"Marks":[200,180,250,145]})
student_data2.set_index("Student_id",inplace=True)
print(student_data2)
```

# Write a Pandas program to join the two given dataframes along rows and assign all data. 

```python
result_rows_join = pd.concat([student_data1, student_data2]) 
result_rows_join
```

# Q.2. Write a Pandas program to join the two given dataframes along columns and assign all data. 

```python
pd.merge(student_data1, student_data2, on = "Name", how = "outer")
```

# 4. Write a Pandas program to merge two given dataframes with different columns. 

```python
Data1 =pd.DataFrame( {
    'key1': ['K0', 'K0', 'K1', 'K2'],
    'key2': ['K0', 'K1', 'K0', 'K1'],
    'P': ['P0', 'P1', 'P2', 'P3'],
    'Q': ['Q0', 'Q1', 'Q2', 'Q3']})
Data1
```

```python
Data2 =pd.DataFrame( {
    'key1': ['K0', 'K1', 'K1', 'K2'],
    'key2': ['K0', 'K0', 'K0', 'K0'],
    'R': ['R0', 'R1', 'R2', 'R3'],
    'S': ['S0', 'S1', 'S2', 'S3']})
Data2
```

```python
pd.concat([Data1, Data2], axis=0, ignore_index=True)
```

# 5. Write a Pandas program to join two dataframes using keys from right dataframe only. 

```python
data = pd.merge(Data1, Data2, how='right', on=['key1', 'key2'])
data
```

# 6. Write a Pandas program to join (left join) the two dataframes using keys from left dataframe only. 

```python
data = pd.merge(Data1, Data2, how='left', on=['key1', 'key2'])
data
```
